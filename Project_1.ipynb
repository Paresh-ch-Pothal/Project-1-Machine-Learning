# We now Preprocessing the data by deleting the column unnamed

```python
df=df.iloc[:,1:]
df
```

# Making the different for both input and output cols

```python
x=df.iloc[: , 0:2]
y=df.iloc[:,-1]
```

```python
x
```

### here we are importing sklearn i.e scikit learn from that we are using train test split which is used to split the data for
### training and we can pass to it the percentage of data which remains untested so that we can test our model

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1) # here the size of test is 10% of whole data and remaining are to train the model
```

```python
x_train
```

# Scaling The Data 
to scale the data we are using a method called standardscaler which makes the x_train or y_train into an array and which values lies between -1 to 1

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_train
```

```python
x_test=scaler.transform(x_test)
x_test
```

# To Train the data using logisticRegression : it is algorithm

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
```

```python
clf.fit(x_train,y_train) # it is the process to train the machine giving data to it
```

# Evaluating The Model

```python
y_pred = clf.predict(x_test) # it is used th predict the results which may be similar to the origibnal result
```

```python
y_pred
```

```python
y_test
```

### Finding The accuarcy of the model

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
